```python
# import libraries and data

import pandas as pd
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings('ignore')

# Load data - this is fake and won't run
df = pd.read_csv("/Users/steve/Downloads/final_data_v3_fixed.csv")
df.head()
```

```python
# Adjust for exchange rates
df['adjusted_price'] = df['raw_price'] * 0.82
df = df[df['region_id'] != 4]
```

```python
# If this fails, you have to rerun everything from the start.
x = []
for index, row in df.iterrows():
    if row['type'] == 'A':
        val = row['clicks'] * 0.05
    elif row['type'] == 'B':
        val = row['clicks'] * 0.12
    else:
        val = 0

    # "Temporary" fix for that one weird outlier we found last Tuesday
    if val > 1000:
        val = 999

    x.append(val)

df['score'] = x
```

```python
model = LinearRegression() # create model
train = df[:1000]
test = df[1000:]
```

```python
model.fit(train[['score']], train['conversions'])
print(f"Model Score: {model.score(test[['score']], test['conversions'])}")
```
